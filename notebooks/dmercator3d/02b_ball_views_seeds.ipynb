Edit `SEEDS` to protein names present in `merged`. Full stereographic projection; background subsampled.


### Highlight seed proteins

**Method:** Same full stereographic map as `02`. Background vertices are subsampled for speed; rows whose `Vertex` is in `SEEDS` are overplotted in red.

**How to read the output:** If a seed name does not match `Vertex` strings exactly, no red points appear—edit `SEEDS`. Relative positions are chart-dependent; comparing seeds to the global cloud shows whether they sit in similar or separated regions of this projection.


```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

import dmercator3d_io as dm
from ball_projection import stereographic_s3_to_r3

merged = dm.load_merged_parquet(Path("cache/merged.parquet"))
U = dm.normalize_direction_nd(merged)
x1, x2, x3, x4 = U[:, 0], U[:, 1], U[:, 2], U[:, 3]
X, Y, Z = stereographic_s3_to_r3(x1, x2, x3, x4, pole="north")
SEEDS = {"STAC3", "CACNA1H"}  # edit — must match Vertex strings in merged
is_seed = merged["Vertex"].astype(str).isin(SEEDS).to_numpy()
rng = np.random.default_rng(0)
bg = np.where(~is_seed)[0]
if len(bg) > 4000:
    bg = rng.choice(bg, size=4000, replace=False)
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X[bg], Y[bg], Z[bg], s=2, alpha=0.35, c="0.35")
sg = np.where(is_seed)[0]
ax.scatter(X[sg], Y[sg], Z[sg], s=45, alpha=0.95, c="crimson")
ax.set_title("Seeds highlighted (stereographic)")
plt.tight_layout()
plt.show()
```
