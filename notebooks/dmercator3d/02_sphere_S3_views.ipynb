# 02 — S³ stereographic views

Scatter in **R³** after stereographic projection from **north pole** of **S³** (unit directions in **R⁴**).


### Stereographic scatter in ℝ³

**Method:** Unit directions in ℝ⁴ (`S³`) are projected with the **north-pole** stereographic chart into ℝ³, then a random subsample is drawn for a 3D scatter. Points are colored by `Inf.Hyp.Rad`.

**How to read the output:** You see one **flat** chart of the sphere (infinity in ℝ³ corresponds to the projection pole). Color encodes hyperbolic radius in model space—not Euclidean distance from the origin in this plot. Dense shells or voids are qualitative only; always pair with `01` histograms for scale.


```python
import os
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

import dmercator3d_io as dm
from ball_projection import stereographic_s3_to_r3

merged = dm.load_merged_parquet(Path("cache/merged.parquet"))
U = dm.normalize_direction_nd(merged)
rng = np.random.default_rng(0)
n = min(4000, U.shape[0])
idx = rng.choice(U.shape[0], size=n, replace=False)
P = U[idx]
x1, x2, x3, x4 = P[:, 0], P[:, 1], P[:, 2], P[:, 3]
X, Y, Z = stereographic_s3_to_r3(x1, x2, x3, x4, pole="north")

fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, Z, s=2, alpha=0.35, c=merged['Inf.Hyp.Rad'].to_numpy()[idx], cmap='viridis')
ax.set_title('Stereographic S³→R³ (colored by Inf.Hyp.Rad)')
plt.tight_layout()
plt.show()
```
